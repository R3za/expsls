```python
from haven import haven_jupyter as hj
from haven import haven_results as hr
from haven import haven_utils as hu
import pprint
import numpy as np

# path to where the experiments are saved
savedir_base = './results'
exp_config_fname = './exp_configs_temp.py'
```

```python
def form_exp_list(exp_groups, exp_config_fname):
    exp_list = []
    for e in exp_groups:
        exp_list += hu.load_py(exp_config_fname).EXP_GROUPS[e]
    return exp_list


# print(form_exp_list(["exp_mushrooms", "exp_ijcnn","exp_rcv1"],exp_config_fname))


markersize = 8
markevery = 15
linewidth = 1
marker = '*'
styles = {}

styles['EXP_SGD_CNST'] = {'label':'SGD_CNST',  'color':'blue','marker':marker ,
                         'markersize':15, 'marker':marker,
                           'linewidth': 2,
                     'markevery':markevery }

styles['EXP_SGD_DECR'] = {'label':'EXP_CNST_DECR',  'color':'purple','marker':marker ,
                         'markersize':15, 'marker':marker,
                           'linewidth': 2,
                     'markevery':markevery }


styles['EXP_ACC_SGD_CNST'] = {'label':'ACC_CNST',  'color':'orange','marker':marker ,
                         'markersize':15, 'marker':marker,
                           'linewidth': 2,
                     'markevery':markevery }

styles['EXP_ACC_SGD_DECR'] = {'label':'ACC_EXP',  'color':'red','marker':marker ,
                         'markersize':15, 'marker':marker,
                           'linewidth': 2,
                     'markevery':markevery }





styles['EXP_SLS_DECR'] = {'label':'EXP_SLS_DECR',  'color':'green','marker':marker ,
                         'markersize':15, 'marker':marker,
                           'linewidth': 2,
                     'markevery':markevery }

styles['RIT_SGD'] = {'label':'R_SGD',  'color':'black','marker':marker ,
                         'markersize':15, 'marker':marker,
                           'linewidth': 2,
                     'markevery':markevery }



```

# ExpList for Losses

```python

datasets = [ "ijcnn","rcv1", "mushrooms"]
exp_groups = []

for dataset in datasets:
    exp_groups.append("exp_" + dataset)
exp_list = form_exp_list(exp_groups, exp_config_fname)

loss_exp_dict={}
combined_filter_dict={}

for loss_func,loss in [("squared_loss","squared")]:
    filtered_best_exp_list=[]
    for dataset in datasets:
        for run in range(5):
#             rho_filter_list = [        
#                 ({'opt':{'name':'RIT_SGD'},"runs":run,  
#                     "loss_func":loss_func, "dataset":dataset})]
# #                    {'style':styles["RIT_SGD"]})]
#             exp_list_filtered = hr.filter_exp_list(exp_list, filterby_list=rho_filter_list, verbose=0)
#             filtered_best_exp_list+=[hr.get_best_exp_dict(exp_list_filtered, savedir_base, 
#                                                 filterby_list=rho_filter_list,
#                                                 metric='train_loss', metric_agg='min', 
#                                                 return_scores=False, verbose=False)]
            for alpha_t in ["CNST","DECR"]: 
                rho_filter_list = [        
                    ({'opt':{'name':'Exp_ACC_SGD','alpha_t':alpha_t },"runs":run,  
                        "loss_func":loss_func, "dataset":dataset})]
#                        {'style':styles["Exp_ACC_SGD_"+alpha_t]})]

                exp_list_filtered = hr.filter_exp_list(exp_list, filterby_list=rho_filter_list, verbose=0)
                filtered_best_exp_list+=[hr.get_best_exp_dict(exp_list_filtered, savedir_base, 
                                                filterby_list=rho_filter_list,
                                                metric='train_loss', metric_agg='min', 
                                                return_scores=False, verbose=False)]

#     rfilteres=[({'opt':{'name':'EXP_SGD'},"loss_func":loss_func})] 
#     rest_exp=hr.filter_exp_list(exp_list, filterby_list=rfilteres, verbose=0)
#     # print(len(rest_exp))        
#     filtered_best_exp_list+=rest_exp

    loss_exp_dict[loss_func]=filtered_best_exp_list 
    
print(len(loss_exp_dict["squared_loss"])   )

for loss_func,loss in [("squared_loss","squared"),("logistic_loss","logistic")]:
    filtered_best_exp_list=[]
    for dataset in datasets:
#             filtered_best_exp_list += [        
#                 ({'opt':{'name':'RIT_SGD'},  
#                     "loss_func":loss_func, "dataset":dataset},
#                     {'style':styles["RIT_SGD"]})]
            
            for alpha_t in ["CNST","DECR"]: 
                
                filtered_best_exp_list += [        
                    ({'opt':{'name':'Exp_ACC_SGD','alpha_t':alpha_t}, 
                        "loss_func":loss_func, "dataset":dataset},
                       {'style':styles["EXP_ACC_SGD_"+alpha_t]})]
                
                
                
#                 filtered_best_exp_list += [        
#                     ({'opt':{'name':'EXP_SGD','alpha_t':alpha_t, "is_sls":False },  
#                         "loss_func":loss_func, "dataset":dataset},
#                        {'style':styles["EXP_SGD_"+alpha_t]})]
            
#             filtered_best_exp_list +=[        
#                     ({'opt':{'name':'EXP_SGD',"is_sls":True },  
#                         "loss_func":loss_func, "dataset":dataset},
#                        {'style':styles["EXP_SLS_DECR"]})]
             
   
    combined_filter_dict[loss_func]=filtered_best_exp_list 
# print(combined_filter_dict["squared_loss"])
# print(loss_exp_dict["squared_loss"])
print(len(hr.filter_exp_list(loss_exp_dict["squared_loss"], filterby_list=combined_filter_dict["squared_loss"], verbose=0)  )  )
    
```

# Squared Loss

```python
show_legend_all=True
rm = hr.ResultManager(
                    exp_list= loss_exp_dict["squared_loss"],
                    savedir_base=savedir_base,
                    filterby_list=combined_filter_dict["squared_loss"],
                    verbose= 0       
                     )


rm.get_plot_all(
    avg_across='runs',
    plot_median=True,
    order='metrics_by_groups',
    show_legend_all=show_legend_all,
    legend_last_row_only=True,
    y_metric_list =  ['grad_norm'], 
    x_metric='itr',
    legend_list=['opt.name'], 
    title_list = ['dataset'], 
    groupby_list = ['dataset', 'model'],
    log_metric_list = ["grad_norm"],
    legend_fontsize=16,
     x_fontsize=16,
     y_fontsize=16,
     xtick_fontsize=16,
     ytick_fontsize=16,
     title_fontsize=18,

    map_ylabel_list=[{'train_loss':'Train loss'}, 
                     {'val_acc':'Validation accuracy'},
                     {'grad_norm': 'Gradient Norm'},
                     {'n_grad_evals': 'Number of stochastic gradient evaluations'}],
    map_xlabel_list=[{'itr': '(Gradient evaluations) / 1000'}],
                figsize=(19,5),
    plot_confidence=True,
#     legend_kwargs={'bbox_to_anchor':[-1.50, -0.26], 'ncol':3, 'loc':'center'}
            )
```
